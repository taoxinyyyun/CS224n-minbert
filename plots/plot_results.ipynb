```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
def plot_results(results, fig_path):
    '''
    plot train loss, train acc and dev acc on two axis for one experiment
    '''
    plt.tight_layout()
    plt.style.use('seaborn')
    # create figure and axes
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 4))

    # plot on the first axis
    ax1.plot(results['epoch'], results['sst_train_loss'])
    ax1.plot(results['epoch'], results['para_train_loss'])
    ax1.plot(results['epoch'], results['sts_train_loss'])
    ax1.set_title('Train loss')

    # plot on the second axis
    ax2.plot(results['epoch'], results['sst_train_acc'])
    ax2.plot(results['epoch'], results['para_train_acc'])
    ax2.plot(results['epoch'], results['sts_train_corr'])
    ax2.set_title('Train performance')

    # plot on the third axis
    ax3.plot(results['epoch'], results['sst_dev_acc'], label='SST')
    ax3.plot(results['epoch'], results['para_dev_acc'], label='Quora')
    ax3.plot(results['epoch'], results['sts_dev_corr'], label='STS')
    ax3.set_title('Dev performance')
    
    fig.supxlabel('Epoch')
    fig.legend()
    
    # display the figure
    plt.savefig(fig_path)
    plt.show()
```

```python
plt.rcParams['figure.dpi'] = 150
```

```python
seq_nopal_results = pd.read_excel('results.xlsx', sheet_name='seq_nopal')
seq_nopal_fig_path = 'seq_nopal.png'
```

```python
plot_results(seq_nopal_results, seq_nopal_fig_path)
```
